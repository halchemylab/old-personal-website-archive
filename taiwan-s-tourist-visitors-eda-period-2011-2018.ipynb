```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load dataset into dataframe
df = pd.read_csv('/kaggle/input/taiwan-visitor-arrivals-by-residence/residence.csv')
```

```python
# Select the data to get only visitors from Mainland China
df_china = df[df['Residence'] == "Mainland China"]
df_china = df_china[['Total', 'Period']]

# Sort the dataframe by period order ascending
df_china = df_china.sort_values(by='Period', ascending=True)

# Create the line plot using Seaborn
sns.lineplot(
    data=df_china,
    x="Period",
    y="Total"
)

# Customize the plot
plt.title("Taiwan's Tourist Visitors from Mainland China (2011-2018)")
plt.ylabel("Visitors")
plt.xlabel(None)

year_ticks = df_china['Period'].loc[df_china['Period'].str.endswith("-01")]
plt.xticks(year_ticks, year_ticks.str[:4])  # Slice the strings to remove "-01"

# Show the plot
plt.show()
```

```python
# Select the data to get only visitors from Mainland China
df_japan = df[df['Residence'] == "Japan"]
df_japan = df_japan[['Total', 'Period']]

# Sort the dataframe by period order ascending
df_japan = df_japan.sort_values(by='Period', ascending=True)

# Create the line plot using Seaborn
sns.lineplot(
    data=df_japan,
    x="Period",
    y="Total"
)

# Customize the plot
plt.title("Taiwan's Tourist Visitors from Japan (2011-2018)")
plt.ylabel("Visitors")
plt.xlabel(None)

year_ticks = df_japan['Period'].loc[df_japan['Period'].str.endswith("-01")]
plt.xticks(year_ticks, year_ticks.str[:4])  # Slice the strings to remove "-01"

# Show the plot
plt.show()
```

```python
# Create the line plot using Seaborn for China
sns.lineplot(
    data=df_china,
    x="Period",
    y="Total",
    label="China",
    color="red"
)

# Create the line plot using Seaborn for Japan
sns.lineplot(
    data=df_japan,
    x="Period",
    y="Total",
    label="Japan",
    color="blue"
)

# Create the line plot using Seaborn for Korea
sns.lineplot(
    data=df_korea,
    x="Period",
    y="Total",
    label="Korea",
    color="green"
)

# Create the line plot using Seaborn for USA
sns.lineplot(
    data=df_usa,
    x="Period",
    y="Total",
    label="USA",
    color="orange"
)

# Customize the plot
plt.title("Taiwan Tourist Visitors from China, Japan, Korea, and USA (2011-2018)")
plt.ylabel("Visitors")
plt.xlabel(None)

# Set ticks for years
year_ticks = df_china['Period'].loc[df_china['Period'].str.endswith("-01")]
plt.xticks(year_ticks, year_ticks.str[:4])

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
